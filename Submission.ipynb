RFM Analisis untuk Registered User berdasarkan bulan

```python
datetime_columns = ["dteday"]
 
for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])

rfm_df = day_df.groupby(by="mnth", as_index=False).agg({
    "dteday": "max", # mengambil tanggal order terakhir
    "registered": "sum", # menghitung jumlah order
    "cnt": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["month", "max_order_timestamp", "frequency", "monetary"]
 
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = day_df["dteday"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.head(13)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]
 
sns.barplot(y="recency", x="month", data=rfm_df.sort_values(by="recency", ascending=False).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
 
sns.barplot(y="frequency", x="month", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
 
sns.barplot(y="monetary", x="month", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
 
plt.suptitle("Registered Customer on Monthly Summary Based on RFM Parameters", fontsize=20)
plt.show()
```

RFM Analisis untuk Casual User berdasarkan bulan

```python
datetime_columns = ["dteday"]
 
for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])

rfm_dfs = day_df.groupby(by="mnth", as_index=False).agg({
    "dteday": "max", # mengambil tanggal order terakhir
    "casual": "sum", # menghitung jumlah order
    "cnt": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_dfs.columns = ["months", "max_order_timestamps", "frequencys", "monetarys"]
 
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_dfs["max_order_timestamps"] = rfm_dfs["max_order_timestamps"].dt.date
recent_dates = day_df["dteday"].dt.date.max()
rfm_dfs["recencys"] = rfm_dfs["max_order_timestamps"].apply(lambda x: (recent_dates - x).days)

rfm_dfs.head(13)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))
 
colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]
 
sns.barplot(y="recencys", x="months", data=rfm_dfs.sort_values(by="recencys", ascending=False).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
 
sns.barplot(y="frequencys", x="months", data=rfm_dfs.sort_values(by="frequencys", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
 
sns.barplot(y="monetarys", x="months", data=rfm_dfs.sort_values(by="monetarys", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
 
plt.suptitle("Casual Customer on Monthly Summary Based on RFM Parameters", fontsize=20)
plt.show()
```

1. Bagaimana Cuaca Mempengaruhi Aktivitas Rental Sepeda?

Secara general tidak ada perbedaan yang cukup mencolok antara casual user dengan registered user. Mereka lebih sedikit menggunakan sepeda ketika cuaca berada pada kategori 3 yaitu Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds.

Pada cuaca seperti itu jumlah registered user yang menggunakan sepeda lebih banyak. Kemungkinan besar mereka menggunakan sepeda untuk aktivitas harian seperti bekerja.

Frekuensi penggunaan sepeda menurun baik registered user maupun casual user pada periode Desember sampai Februari dikarenakan pada periode tersebut adalah musim dingin

2. Apakah ada perbedaan perilaku harian antara Casual User dan Registered User?

Registered user lebih cenderung menggunakan sepeda saat hari kantor terlihat pada grafik sebelah kanan yang cenderung lebih banyak penyewaan pada periode weekday dan lebih banyak digunakaan saat memasuki jam kerja dan pulang kerja. Sebaliknya, casual user lebih cenderung menggunakan sepeda saat hari libur dan pada waktu untuk berekreasi sekitar jam 10 - 17.

Jika dilihat sekilas kuantitas penggunaan sepeda saat hari libur masih didominasi oleh registered user dan tentunya ada lonjakan yang cukup signifikan pada saat hari libur dan juga rentang waktu jam 17 sampai 18.
